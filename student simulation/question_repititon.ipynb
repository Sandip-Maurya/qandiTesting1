```python
# find repitive questions between 2 files
def find_repititive_questions(file1,file2):
    try:
        df1 = pd.read_csv(file1)
        df2 = pd.read_csv(file2)
        return df1.merge(df2,on='q_id').shape[0]    
    except:
        pass
```

```python
find_repititive_questions('botTestOutput\M_Area Bounded by Cures.csv','uniqtd@gmail.com-1\M_Area Bounded by Curves.csv')
```

```python
def bulk_repitition_check(folder1,folder2):
    for file1 in os.listdir(folder1):
        if file1[-3:] == 'csv':
            if file1 in os.listdir(folder2):
                print(file1," : ",end="")
                print(find_repititive_questions(os.path.join(folder1,file1) , os.path.join(folder2,file1)))
```

```python
bulk_repitition_check('botTestOutput','uniqtd@gmail.com-1')
```
