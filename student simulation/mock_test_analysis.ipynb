```python
files = os.listdir('mock_test')
```

```python
# find repitive questions between 2 files
def find_repititive_questions(file1,file2):
    try:
        df1 = pd.read_csv(file1)
        df2 = pd.read_csv(file2)
        df_merged = df1.merge(df2,on='q_id')['q_id']
        #print(df_merged)
        return df_merged.shape[0]    
    except Exception as e:
        print(f'something went wrong : {e}')
```

```python
f1 = os.path.join(os.getcwd(),'mock_test',files[2])
f2 = os.path.join(os.getcwd(),'mock_test',files[4])

find_repititive_questions(f1,f2)
```

```python
n = len(os.listdir('mock_test'))
for i in range(n):
    for j in range(i+1,n):
        try:
            f1 = os.path.join(os.getcwd(),'mock_test',files[i])
            f2 = os.path.join(os.getcwd(),'mock_test',files[j])

            common_ques = find_repititive_questions(f1,f2)

            print(f'{i} {j} : {common_ques}')
        except Exception as e:
            print(e)
        
```
