```python
import re
import pandas as pd
df = pd.read_excel('testing_for_AdaptiveAPI.xlsx')
# f = open(,encoding="utf8")
# x = f.readlines()
# x[:100]
df[:20]
```

```python
df = df.fillna('')
df[:4]
```

```python
df = df.astype('str')
df.dtypes
```

```python
df['combined'] = df[df.columns.values].T.agg(','.join)
df['combined']
```

```python
new_df = pd.DataFrame(columns = ['q_id', 'corr_opt', 'diff_lev', 'top_id', 'chap_id'])
pattern = r'(\d+),\{(\d)":".*?"\}",(\d+),(\d+),(\d+)'
n = len(df)
# n = 50
for i in range(n):
    data = re.findall(pattern, df.loc[i,'combined'])
    if data:
        new_df.loc[i,:] = data[0]
    if i%1000 == 0:
        print(f'Data extracted for total {i+1} questions')
new_df.to_csv('q_data.csv')
print('done')
```
